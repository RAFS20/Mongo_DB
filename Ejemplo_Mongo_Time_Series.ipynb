````python
import yfinance as yf
import matplotlib.pyplot as plt

# Crear un objeto ticker para la acción que deseas analizar
msft = yf.Ticker("MSFT")

# Obtener datos históricos para la acción
historical_data = msft.history(period="5y")

# Obtener información en tiempo real
real_time_data = msft.info

# Obtener información de dividendos
dividend_data = msft.dividends

# Crear un gráfico de precios de acciones
historical_data['Close'].plot()
plt.title('Precio de cierre de Microsoft en los últimos 5 años')
plt.xlabel('Fecha')
plt.ylabel('Precio de cierre')
plt.show()

# Imprimir información en tiempo real
print("Información en tiempo real:")
for key, value in real_time_data.items():
    print(f"{key}: {value}")

# Imprimir información de dividendos
print("\nInformación de dividendos:")
print(dividend_data)
````

````python
# Calcular el promedio móvil de 50 días
historical_data['SMA_50'] = historical_data['Close'].rolling(window=50).mean()

# Calcular el RSI (Índice de Fuerza Relativa)
def calculate_rsi(data, period=14):
    delta = data['Close'].diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

historical_data['RSI'] = calculate_rsi(historical_data)
# Crear un gráfico de precios de acciones
historical_data['Close'].plot(label='Precio de cierre')

# Crear un gráfico del promedio móvil de 50 días
historical_data['SMA_50'].plot(label='SMA-50')

plt.title('Precio de cierre y SMA-50 de Microsoft en los últimos 5 años')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend(loc='upper left')

# Crear un gráfico del RSI
plt.figure(figsize=(12, 6))
plt.plot(historical_data.index, historical_data['RSI'], label='RSI', color='purple')
plt.axhline(y=70, color='red', linestyle='--', label='Overbought (70)')
plt.axhline(y=30, color='green', linestyle='--', label='Oversold (30)')
plt.fill_between(historical_data.index, y1=30, y2=70, color='yellow', alpha=0.2)
plt.title('RSI de Microsoft en los últimos 5 años')
plt.xlabel('Fecha')
plt.ylabel('RSI')
plt.legend()
plt.show()
````

# Índice de Fuerza Relativa (RSI, por sus siglas en inglés).

El RSI es un oscilador de momento que mide la velocidad y cambio de los movimientos de precios de un activo y se utiliza para identificar condiciones de sobrecompra o sobreventa en el mercado de un activo.

La gráfica muestra el RSI de Microsoft durante los últimos 5 años. El RSI oscila entre 0 y 100 y se puede observar en la línea púrpura. Las líneas horizontalmente trazadas representan los niveles típicos de sobrecompra y sobreventa. La línea roja punteada que está marcada como "Overbought (70)" indica el nivel de sobrecompra, que es cuando el RSI está por encima de 70. Esto sugiere que el activo podría estar sobrevalorado y que podría haber una corrección o cambio de tendencia a la baja. La línea verde punteada que está marcada como "Oversold (30)" indica el nivel de sobreventa, que es cuando el RSI está por debajo de 30. Esto podría indicar que el activo está infravalorado y que podría haber un cambio de tendencia al alza.

La zona sombreada en amarillo sugiere períodos donde el RSI ha estado entre estos dos extremos, lo que se consideraría un rango "normal" de operación. La fecha está marcada en el eje horizontal, y el valor del RSI está marcado en el eje vertical.

El análisis del RSI es comúnmente utilizado por los operadores de mercado para tomar decisiones de compra y venta basadas en las condiciones de sobrecompra y sobreventa, así como para predecir posibles reversiones de la tendencia. Sin embargo, es importante señalar que el RSI es solo una herramienta de análisis y se debe usar en conjunto con otras formas de análisis técnico y fundamental para tomar decisiones de inversión más informadas.

def upload_model_to_mongodb(model, collection_name):
    # Conectar a MongoDB
    client = connect_to_mongo_db()
    db = client.get_database()

    # Convertir el modelo a formato JSON
    model_json = model.to_json()

    # Actualizar el modelo en la colección especificada o insertarlo si no existe

# Red Neuronal Recurrente (RNN) Simple

Una Red Neuronal Recurrente (RNN) es un tipo de red neuronal que se utiliza comúnmente para modelar datos secuenciales, como series temporales, texto y audio. La principal característica de una RNN es su capacidad para mantener y utilizar información de estados anteriores en la secuencia para hacer predicciones en el presente. A diferencia de las redes neuronales convolucionales (CNN), que son ideales para datos espaciales, las RNN son adecuadas para datos secuenciales.

## Estructura de una RNN Simple

La estructura básica de una RNN simple consta de una capa de entrada, una capa oculta y una capa de salida. La capa oculta es la que permite que la red mantenga información sobre estados anteriores a medida que procesa la secuencia. A continuación, describiremos matemáticamente esta estructura.

### Capa de Entrada
La capa de entrada en el tiempo $t$ está representada por el vector $X_t$, que contiene las características de entrada en ese momento. Supongamos que $X_t$ es un vector de dimensión $d$, es decir, $X_t \in \mathbb{R}^d$. Cada componente de $X_t$ corresponde a una característica de entrada.

### Capa Oculta
La capa oculta en el tiempo $t$ está representada por el vector de estado oculto $H_t$. La función matemática que describe cómo se calcula $H_t$ a partir de $X_t$, $H_{t-1}$ (estado oculto anterior), y los pesos de la red se define mediante la siguiente ecuación:

$$
H_t = \sigma(W_{hx}X_t + W_{hh}H_{t-1} + b_h)
$$

Donde:

- $H_t \in \mathbb{R}^h$ es el estado oculto en el tiempo $t$, con $h$ siendo la dimensión del estado oculto.
- $\sigma$ es una función de activación como la función sigmoide o la función tangente hiperbólica.
- $W_{hx} \in \mathbb{R}^{h \times d}$ es la matriz de pesos que conecta la capa de entrada ($X_t$) con la capa oculta ($H_t$).
- $W_{hh} \in \mathbb{R}^{h \times h}$ es la matriz de pesos que conecta la capa oculta anterior ($H_{t-1}$) con la capa oculta actual ($H_t$).
- $b_h \in \mathbb{R}^h$ es el vector de sesgo de la capa oculta.

### Capa de Salida
La capa de salida en el tiempo $t$ está representada por el vector $Y_t$. La función matemática que describe cómo se calcula $Y_t$ a partir de $H_t$ y los pesos de la red se define mediante la siguiente ecuación:

$$
Y_t = \sigma(W_{yh}H_t + b_y)
$$

Donde:

- $Y_t \in \mathbb{R}^o$ es el vector de salida en el tiempo $t$, con $o$ siendo la dimensión de la salida.
- $\sigma$ es una función de activación aplicada a la capa de salida.
- $W_{yh} \in \mathbb{R}^{o \times h}$ es la matriz de pesos que conecta la capa oculta ($H_t$) con la capa de salida ($Y_t$).
- $b_y \in \mathbb{R}^o$ es el vector de sesgo de la capa de salida.

Estas ecuaciones describen matemáticamente cómo se calcula el estado oculto $H_t$ y la salida $Y_t$ en cada paso de tiempo $t$ en una RNN simple. La red puede mantener información sobre estados anteriores a través de la recursión de la ecuación del estado oculto, lo que la hace adecuada para el procesamiento de secuencias temporales.

## Matemáticas detrás de una RNN Simple

Ahora, exploremos las fórmulas matemáticas que describen cómo funciona una RNN simple en cada paso de tiempo. Para simplificar, asumiremos que tanto la capa de entrada como la capa oculta tienen una sola neurona, pero estas fórmulas pueden generalizarse para redes más grandes.

### 1. Propagación hacia adelante (Forward Propagation)

En la propagación hacia adelante, calculamos el estado oculto en el tiempo actual ($H_t$) y la salida en el tiempo actual ($Y_t$) utilizando las siguientes ecuaciones:

#### Cálculo del estado oculto ($H_t$) en el tiempo $t$:

$$
H_t = \sigma(W_{hx}X_t + W_{hh}H_{t-1} + b_h)
$$

Donde:

- $X_t$ es el vector de entrada en el tiempo $t$.
- $H_{t-1}$ es el estado oculto anterior en el tiempo $t-1$ (inicializado en $H_0$ para $t=0$).
- $W_{hx}$ es la matriz de pesos que conecta la capa de entrada ($X_t$) con la capa oculta ($H_t$).
- $W_{hh}$ es la matriz de pesos que conecta la capa oculta anterior ($H_{t-1}$) con la capa oculta actual ($H_t$).
- $b_h$ es el vector de sesgo de la capa oculta.
- $\sigma$ es una función de activación, comúnmente la función sigmoide o la función tangente hiperbólica.

#### Cálculo de la salida ($Y_t$) en el tiempo $t$:

$$
Y_t = \sigma(W_{yh}H_t + b_y)
$$

Donde:

- $H_t$ es el estado oculto en el tiempo $t$.
- $W_{yh}$ es la matriz de pesos que conecta la capa oculta ($H_t$) con la capa de salida ($Y_t$).
- $b_y$ es el vector de sesgo de la capa de salida.

### 2. Retropropagación del Error (Backpropagation)

Para entrenar la RNN y ajustar sus pesos, utilizamos el algoritmo de retropropagación del error (backpropagation). Este algoritmo se basa en la derivación de las funciones de pérdida con respecto a los pesos de la red.

#### Pérdida en el tiempo $t$:

Supongamos que tenemos una función de pérdida $\mathcal{L}$ que evalúa cuán cerca están nuestras predicciones ($Y_t$) de los valores reales ($\hat{Y}_t$) en el tiempo $t$. La pérdida en el tiempo $t$ se denota como $\mathcal{L}_t$ y puede variar según el problema (por ejemplo, error cuadrático medio, entropía cruzada, etc.).

#### Derivadas parciales:

Para ajustar los pesos, necesitamos calcular las derivadas parciales de la pérdida con respecto a los pesos de la red. Usaremos la regla de la cadena para hacer esto. Las derivadas parciales se calcularán en relación con los pesos $W_{hx}$, $W_{hh}$, $b_h$, $W_{yh}$ y $b_y$.

La derivada parcial de $\mathcal{L}_t$ con respecto a $W_{yh}$ es:

$$
\frac{\partial \mathcal{L}_t}{\partial W_{yh}} = \frac{\partial \mathcal{L}_t}{\partial Y_t} \cdot \frac{\partial Y_t}{\partial W_{yh}}
$$

La derivada parcial de $\mathcal{L}_t$ con respecto a $b_y$ es:

$$
\frac{\partial \mathcal{L}_t}{\partial b_y} = \frac{\partial \mathcal{L}_t}{\partial Y_t} \cdot \frac{\partial Y_t}{\partial b_y}
$$

La derivada parcial de $\mathcal{L}_t$ con respecto a $H_t$ se calcula como:

$$
\frac{\partial \mathcal{L}_t}{\partial H_t} = \frac{\partial \mathcal{L}_t}{\partial Y_t} \cdot \frac{\partial Y_t}{\partial H_t}
$$

La derivada parcial de $\mathcal{L}_t$ con respecto a $W_{hx}$ es:

$$
\frac{\partial \mathcal{L}_t}{\partial W_{hx}} = \frac{\partial \mathcal{L}_t}{\partial H_t} \cdot \frac{\partial H_t}{\partial W_{hx}}
$$

La derivada parcial de $\mathcal{L}_t$ con respecto a $W_{hh}$ es:

$$
\frac{\partial \mathcal{L}_t}{\partial W_{hh}} = \frac{\partial \mathcal{L}_t}{\partial H_t} \cdot \frac{\partial H_t}{\partial W_{hh}}
$$

La derivada parcial de $\mathcal{L}_t$ con respecto a $b_h$ es:

$$
\frac{\partial \mathcal{L}_t}{\partial b_h} = \frac{\partial \mathcal{L}_t}{\partial H_t} \cdot \frac{\partial H_t}{\partial b_h}
$$

### 3. Actualización de los Pesos

Una vez que hemos calculado las derivadas parciales, podemos usar algoritmos de optimización como el descenso de gradiente estocástico (SGD) o el optimizador Adam para actualizar los pesos de la red en cada paso de tiempo.

### Resumen de las Fórmulas

Para resumir, las fórmulas matemáticas clave de una RNN simple para series temporales son:

**Propagación hacia adelante:**

- Cálculo del estado oculto en el tiempo $t$:
  $$
  H_t = \sigma(W_{hx}X_t + W_{hh}H_{t-1} + b_h)
  $$

- Cálculo de la salida en el tiempo $t$:
  $$
  Y_t = \sigma(W_{yh}H_t + b_y)
  $$

**Retropropagación del error:**

- Derivada parcial de la pérdida en el tiempo $t$ con respecto a $W_{yh}$:
  $$
  \frac{\partial \mathcal{L}_t}{\partial W_{yh}} = \frac{\partial \mathcal{L}_t}{\partial Y_t} \cdot \frac{\partial Y_t}{\partial W_{yh}}
  $$

- Derivada parcial de la pérdida en el tiempo $t$ con respecto a $b_y$:
  $$
  \frac{\partial \mathcal{L}_t}{\partial b_y} = \frac{\partial \mathcal{L}_t}{\partial Y_t} \cdot \frac{\partial Y_t}{\partial b_y}
  $$

- Derivada parcial de la pérdida en el tiempo $t$ con respecto a $H_t$:
  $$
  \frac{\partial \mathcal{L}_t}{\partial H_t} = \frac{\partial \mathcal{L}_t}{\partial Y_t} \cdot \frac{\partial Y_t}{\partial H_t}
  $$

- Derivada parcial de la pérdida en el tiempo $t$ con respecto a $W_{hx}$:
  $$
  \frac{\partial \mathcal{L}_t}{\partial W_{hx}} = \frac{\partial \mathcal{L}_t}{\partial H_t} \cdot \frac{\partial H_t}{\partial W_{hx}}
  $$

- Derivada parcial de la pérdida en el tiempo $t$ con respecto a $W_{hh}$:
  $$
  \frac{\partial \mathcal{L}_t}{\partial W_{hh}} = \frac{\partial \mathcal{L}_t}{\partial H_t} \cdot \frac{\partial H_t}{\partial W_{hh}}
  $$

- Derivada parcial de la pérdida en el tiempo $t$ con respecto a $b_h$:
  $$
  \frac{\partial \mathcal{L}_t}{\partial b_h} = \frac{\partial \mathcal{L}_t}{\partial H_t} \cdot \frac{\partial H_t}{\partial b_h}
  $$

Estas fórmulas forman la base matemática de una Red Neuronal Recurrente (RNN) simple para el análisis y la predicción de series temporales. Con estas ecuaciones, podemos entrenar y utilizar una RNN para una amplia variedad de aplicaciones relacionadas con datos secuenciales.

# Red Neuronal Recurrente con Unidades de Puerta (GRU)

Las Unidades de Puerta Recurrentes (GRU) son una variante de las redes neuronales recurrentes (RNN) que se han desarrollado para abordar el problema de la desaparición del gradiente y mejorar la capacidad de las RNN para capturar dependencias a largo plazo en secuencias de datos. Las GRU utilizan mecanismos de puertas para controlar el flujo de información en la red, lo que las hace especialmente efectivas en aplicaciones de series temporales.

## Estructura de una GRU

Una GRU consta de tres componentes principales: una celda de memoria, una unidad de reinicio (reset gate), y una unidad de actualización (update gate). La celda de memoria almacena información temporal relevante, mientras que las puertas controlan cuándo y qué información se almacena o se olvida en la celda de memoria.

### Celda de Memoria

La celda de memoria en una GRU en el tiempo $t$ está representada por $C_t$. En cada paso de tiempo, la celda de memoria se actualiza de la siguiente manera:

$$
C_t = (1 - Z_t) \odot C_{t-1} + Z_t \odot \tilde{C}_t
$$

Donde:

- $Z_t$ es la salida de la unidad de actualización en el tiempo $t$, que determina cuánto de la celda de memoria anterior se debe mantener y cuánto se debe actualizar.
- $\odot$ denota la multiplicación elemento por elemento (producto Hadamard).
- $C_{t-1}$ es la celda de memoria anterior en el tiempo $t-1$.
- $\tilde{C}_t$ es el candidato de celda de memoria que se calculará en función de la entrada actual y el estado anterior.

### Unidad de Actualización (Update Gate)

La unidad de actualización en el tiempo $t$ está representada por $Z_t$. Su propósito es decidir cuánta información del estado anterior debe ser olvidada y cuánta información nueva debe ser almacenada en la celda de memoria. La salida de la unidad de actualización se calcula de la siguiente manera:

$$
Z_t = \sigma(W_z \cdot [H_{t-1}, X_t] + b_z)
$$

Donde:

- $\sigma$ es la función de activación sigmoide.
- $W_z$ es una matriz de pesos que se utiliza para calcular la salida de la unidad de actualización.
- $H_{t-1}$ es el estado oculto anterior en el tiempo $t-1$.
- $X_t$ es el vector de entrada en el tiempo $t$.
- $b_z$ es el vector de sesgo para la unidad de actualización.

### Unidad de Reinicio (Reset Gate)

La unidad de reinicio en el tiempo $t$ está representada por $R_t$. Su función es decidir cuánta información del estado anterior se debe olvidar antes de calcular el candidato de celda de memoria $\tilde{C}_t$. La salida de la unidad de reinicio se calcula de la siguiente manera:

$$
R_t = \sigma(W_r \cdot [H_{t-1}, X_t] + b_r)
$$

Donde:

- $\sigma$ es la función de activación sigmoide.
- $W_r$ es una matriz de pesos que se utiliza para calcular la salida de la unidad de reinicio.
- $H_{t-1}$ es el estado oculto anterior en el tiempo $t-1$.
- $X_t$ es el vector de entrada en el tiempo $t$.
- $b_r$ es el vector de sesgo para la unidad de reinicio.

### Candidato de Celda de Memoria

El candidato de celda de memoria $\tilde{C}_t$ se calcula utilizando el resultado de la unidad de reinicio $R_t$, y se utiliza para actualizar la celda de memoria $C_t$. El cálculo se realiza de la siguiente manera:

$$
\tilde{C}_t = \tanh(W_c \cdot [R_t \odot H_{t-1}, X_t] + b_c)
$$

Donde:

- $\tanh$ es la función tangente hiperbólica.
- $W_c$ es una matriz de pesos que se utiliza para calcular el candidato de celda de memoria.
- $R_t$ es la salida de la unidad de reinicio.
- $H_{t-1}$ es el estado oculto anterior en el tiempo $t-1$.
- $X_t$ es el vector de entrada en el tiempo $t$.
- $b_c$ es el vector de sesgo para el candidato de celda de memoria.

### Estado Oculto de Salida

Finalmente, el estado oculto de salida en el tiempo $t$ se calcula utilizando la celda de memoria $C_t$ y la unidad de actualización $Z_t$:

$$
H_t = (1 - Z_t) \odot \tanh(C_t) + Z_t \odot H_{t-1}
$$

Donde:

- $H_t$ es el estado oculto de salida en el tiempo $t$.
- $Z_t$ es la salida de la unidad de actualización.
- $\tanh$ es la función tangente hiperbólica.
- $C_t$ es la celda de memoria en el tiempo $t$.
- $H_{t-1}$ es el estado oculto anterior en el tiempo $t-1$.

## Resumen de las Fórmulas

Las fórmulas matemáticas clave para una Red Neuronal Recurrente con Unidades de Puerta (GRU) son:

- Cálculo de la celda de memoria en el tiempo $t$:
$$
C_t = (1 - Z_t) \odot C_{t-1} + Z_t \odot \tilde{C}_t
$$

- Cálculo de la salida de la unidad de actualización en el tiempo $t$:
$$
Z_t = \sigma(W_z \cdot [H_{t-1}, X_t] + b_z)
$$

- Cálculo de la salida de la unidad de reinicio en el tiempo $t$:


$$
R_t = \sigma(W_r \cdot [H_{t-1}, X_t] + b_r)
$$

- Cálculo del candidato de celda de memoria en el tiempo $t$:
$$
\tilde{C}_t = \tanh(W_c \cdot [R_t \odot H_{t-1}, X_t] + b_c)
$$

- Cálculo del estado oculto de salida en el tiempo $t$:
$$
H_t = (1 - Z_t) \odot \tanh(C_t) + Z_t \odot H_{t-1}
$$

Estas ecuaciones son fundamentales para entender cómo funciona una GRU en el procesamiento de series temporales. Las unidades de puerta (update y reset gates) permiten que la GRU controle el flujo de información y capture dependencias a largo plazo en los datos de manera efectiva.

## Objetivo del Entrenamiento

El objetivo principal del entrenamiento de una GRU es minimizar una función de pérdida que mida la diferencia entre las predicciones de la red y los valores reales en un conjunto de datos de entrenamiento. En el contexto de series temporales, esto implica aprender a predecir los valores futuros en función de los datos históricos. Para lograr esto, se utiliza una función de pérdida apropiada, como el error cuadrático medio (MSE) o la entropía cruzada, según la naturaleza del problema.

## Proceso de Entrenamiento

El proceso de entrenamiento de una GRU implica los siguientes pasos:

1. **Inicialización de Pesos y Sesgos**: En el inicio del entrenamiento, los pesos y sesgos de la GRU se inicializan de manera aleatoria o utilizando estrategias como Xavier o He initialization. Estos valores iniciales determinan cómo la red realiza las predicciones iniciales.

2. **Propagación hacia adelante (Forward Propagation)**: Durante cada paso de tiempo en el conjunto de datos de entrenamiento, se realiza la propagación hacia adelante para calcular las predicciones de la red. Esto implica el cálculo del estado oculto $H_t$ y la predicción $Y_t$ en función de las entradas $X_t$, utilizando las ecuaciones que se describieron previamente.

3. **Cálculo de la Pérdida (Loss)**: Para evaluar qué tan buenas son las predicciones de la red, se calcula una medida de error utilizando una función de pérdida. Esta función compara las predicciones $Y_t$ con los valores reales $\hat{Y}_t$ en el conjunto de datos de entrenamiento. El objetivo es minimizar esta función de pérdida.

   - **Ejemplo con MSE** (Error Cuadrático Medio):
   $$
   \mathcal{L}_t = \frac{1}{2}(Y_t - \hat{Y}_t)^2
   $$
   
   Donde $\mathcal{L}_t$ es la pérdida en el tiempo $t$.

4. **Retropropagación del Error (Backpropagation)**: Para ajustar los pesos y sesgos de la red y minimizar la función de pérdida, se utiliza el algoritmo de retropropagación del error (backpropagation). Este algoritmo calcula las derivadas parciales de la pérdida con respecto a todos los parámetros de la red, incluidos los pesos y los sesgos.

   - **Ejemplo de Derivadas Parciales**:
   Para el peso $W_{yh}$ que conecta la capa oculta con la capa de salida, la derivada parcial de la pérdida con respecto a $W_{yh}$ se calcula mediante la cadena de regla:
   $$
   \frac{\partial \mathcal{L}_t}{\partial W_{yh}} = \frac{\partial \mathcal{L}_t}{\partial Y_t} \cdot \frac{\partial Y_t}{\partial W_{yh}}
   $$
   
   Esto se repite para todos los pesos y sesgos de la red.

5. **Actualización de Pesos (Optimización)**: Una vez que se han calculado las derivadas parciales, se utilizan algoritmos de optimización como el descenso de gradiente estocástico (SGD), Adam, RMSProp, entre otros, para actualizar los pesos y los sesgos de la red en la dirección que minimiza la función de pérdida.

   - **Ejemplo de Actualización de Pesos con SGD**:
   $$
   W_{yh} = W_{yh} - \alpha \frac{\partial \mathcal{L}_t}{\partial W_{yh}}
   $$
   
   Donde $\alpha$ es la tasa de aprendizaje que controla la magnitud de la actualización de los pesos.

6. **Repetición del Proceso**: Los pasos 2 al 5 se repiten durante múltiples épocas (iteraciones completas a través del conjunto de datos de entrenamiento) hasta que la función de pérdida converja o alcance un valor aceptable.

7. **Validación y Evaluación**: Después del entrenamiento, es importante evaluar el rendimiento de la GRU en un conjunto de datos de validación o prueba independiente para asegurarse de que pueda generalizar bien a datos no vistos. Se utilizan métricas de evaluación como el error cuadrático medio (MSE), el error absoluto medio (MAE), o coeficiente de correlación para medir la calidad de las predicciones.

8. **Ajuste de Hiperparámetros**: Durante el proceso de entrenamiento, es posible que sea necesario ajustar los hiperparámetros de la GRU, como la tasa de aprendizaje, el tamaño de la capa oculta, la función de activación, etc., para optimizar el rendimiento del modelo.

Este proceso de entrenamiento iterativo permite que una GRU aprenda patrones en los datos de series temporales y mejore su capacidad para hacer predicciones precisas. La elección adecuada de hiperparámetros, la función de pérdida y la arquitectura de la red son factores críticos para lograr un buen rendimiento en aplicaciones de series temporales.


# Red Neuronal Recurrente de Memoria a Largo Plazo (LSTM)

Las LSTMs son una variante de las redes neuronales recurrentes (RNN) que han demostrado ser efectivas en tareas de predicción de series temporales debido a su capacidad para capturar y recordar relaciones a largo plazo en los datos. A diferencia de las RNN tradicionales, las LSTMs utilizan unidades de memoria especiales para lograr esto. A continuación, describiré matemáticamente las LSTMs y cómo funcionan.

## Estructura de una LSTM

Una LSTM consta de tres puertas (gates) principales que regulan el flujo de información a través de la unidad de memoria: la puerta de olvido (forget gate), la puerta de entrada (input gate), y la puerta de salida (output gate). Estas puertas determinan qué información se debe olvidar, qué información se debe agregar a la memoria y cuándo se debe leer de la memoria.

### Unidad de Memoria

La unidad de memoria en una LSTM en el tiempo $t$ está representada por $C_t$. En cada paso de tiempo, la unidad de memoria se actualiza de la siguiente manera:

$$
C_t = f_t \odot C_{t-1} + i_t \odot \tilde{C}_t
$$

Donde:

- $C_t$ es la unidad de memoria en el tiempo $t$.
- $f_t$ es la salida de la puerta de olvido (forget gate) en el tiempo $t$, que determina cuánto de la memoria anterior se debe mantener y cuánto se debe olvidar.
- $i_t$ es la salida de la puerta de entrada (input gate) en el tiempo $t$, que determina cuánta información nueva se debe agregar a la memoria.
- $\odot$ denota la multiplicación elemento por elemento (producto Hadamard).
- $\tilde{C}_t$ es el candidato de memoria que se calculará en función de la entrada actual y el estado anterior.

### Puerta de Olvido (Forget Gate)

La puerta de olvido en el tiempo $t$ está representada por $f_t$. Su función es decidir cuánta información de la memoria anterior $C_{t-1}$ debe ser olvidada antes de calcular el candidato de memoria $\tilde{C}_t$. La salida de la puerta de olvido se calcula de la siguiente manera:

$$
f_t = \sigma(W_f \cdot [H_{t-1}, X_t] + b_f)
$$

Donde:

- $\sigma$ es la función de activación sigmoide.
- $W_f$ es una matriz de pesos que se utiliza para calcular la salida de la puerta de olvido.
- $H_{t-1}$ es el estado oculto anterior en el tiempo $t-1$.
- $X_t$ es el vector de entrada en el tiempo $t$.
- $b_f$ es el vector de sesgo para la puerta de olvido.

### Puerta de Entrada (Input Gate)

La puerta de entrada en el tiempo $t$ está representada por $i_t$. Su función es decidir cuánta información nueva se debe agregar a la memoria $C_t$. La salida de la puerta de entrada se calcula de la siguiente manera:

$$
i_t = \sigma(W_i \cdot [H_{t-1}, X_t] + b_i)
$$

Donde:

- $\sigma$ es la función de activación sigmoide.
- $W_i$ es una matriz de pesos que se utiliza para calcular la salida de la puerta de entrada.
- $H_{t-1}$ es el estado oculto anterior en el tiempo $t-1$.
- $X_t$ es el vector de entrada en el tiempo $t$.
- $b_i$ es el vector de sesgo para la puerta de entrada.

### Candidato de Memoria

El candidato de memoria $\tilde{C}_t$ se calcula utilizando el resultado de la puerta de entrada $i_t$, y se utiliza para actualizar la memoria $C_t$. El cálculo se realiza de la siguiente manera:

$$
\tilde{C}_t = \tanh(W_C \cdot [H_{t-1}, X_t] + b_C)
$$

Donde:

- $\tanh$ es la función tangente hiperbólica.
- $W_C$ es una matriz de pesos que se utiliza para calcular el candidato de memoria.
- $H_{t-1}$ es el estado oculto anterior en el tiempo $t-1$.
- $X_t$ es el vector de entrada en el tiempo $t$.
- $b_C$ es el vector de sesgo para el candidato de memoria.

### Puerta de Salida (Output Gate)

La puerta de salida en el tiempo $t$ está representada por $o_t$. Su función es decidir cuánta información de la memoria $C_t$ se debe leer y cuánta se debe pasar al estado oculto $H_t$. La salida de la puerta de salida se calcula de la siguiente manera:

$$
o_t = \sigma(W_o \cdot [H_{t-1}, X_t] + b_o)
$$

Donde:

- $\sigma$ es la función de activación sigmoide.
- $W_o$ es una matriz de pesos que se utiliza para calcular la salida de la puerta de salida.
- $H_{t-1}$ es el estado oculto anterior en el tiempo $t-1$.
- $X_t$ es el vector de entrada en el tiempo $t$.
- $b_o$ es el vector de sesgo para la puerta de salida.

### Estado Oculto de Salida

Finalmente, el estado oculto de salida en el tiempo $t$ se calcula utilizando la memoria

 $C_t$ y la puerta de salida $o_t$:

$$
H_t = o_t \odot \tanh(C_t)
$$

Donde:

- $H_t$ es el estado oculto de salida en el tiempo $t$.
- $o_t$ es la salida de la puerta de salida.
- $\tanh$ es la función tangente hiperbólica.
- $C_t$ es la unidad de memoria en el tiempo $t$.

## Resumen de las Fórmulas

Las fórmulas matemáticas clave para una Red Neuronal Recurrente de Memoria a Largo Plazo (LSTM) son:

- Cálculo de la unidad de memoria en el tiempo $t$:
$$
C_t = f_t \odot C_{t-1} + i_t \odot \tilde{C}_t
$$

- Cálculo de la puerta de olvido en el tiempo $t$:
$$
f_t = \sigma(W_f \cdot [H_{t-1}, X_t] + b_f)
$$

- Cálculo de la puerta de entrada en el tiempo $t$:
$$
i_t = \sigma(W_i \cdot [H_{t-1}, X_t] + b_i)
$$

- Cálculo del candidato de memoria en el tiempo $t$:
$$
\tilde{C}_t = \tanh(W_C \cdot [H_{t-1}, X_t] + b_C)
$$

- Cálculo de la puerta de salida en el tiempo $t$:
$$
o_t = \sigma(W_o \cdot [H_{t-1}, X_t] + b_o)
$$

- Cálculo del estado oculto de salida en el tiempo $t$:
$$
H_t = o_t \odot \tanh(C_t)
$$

Estas ecuaciones describen cómo funciona una LSTM y cómo se calcula el estado de la memoria y el estado oculto en cada paso de tiempo. La capacidad de recordar información a largo plazo y aprender dependencias complejas en las secuencias de datos hace que las LSTMs sean ampliamente utilizadas en aplicaciones de series temporales y procesamiento de lenguaje natural.

## Objetivo del Entrenamiento

El objetivo del entrenamiento es ajustar los parámetros de la LSTM, que incluyen pesos ($W_f$, $W_i$, $W_C$, $W_o$) y sesgos ($b_f$, $b_i$, $b_C$, $b_o$), de manera que la función de pérdida se minimice. La función de pérdida mide la discrepancia entre las predicciones del modelo y los valores reales. En el contexto de series temporales, las pérdidas típicas incluyen el Error Cuadrático Medio (MSE) o la Entropía Cruzada Categórica (para problemas de clasificación).

## Proceso de Entrenamiento

El proceso de entrenamiento de una LSTM implica los siguientes pasos:

1. **Inicialización de Pesos y Sesgos**: Al inicio del entrenamiento, los pesos y sesgos de la LSTM se inicializan típicamente de manera aleatoria o utilizando estrategias de inicialización como Xavier o He initialization.

2. **Propagación hacia adelante (Forward Propagation)**: Durante cada paso de tiempo en el conjunto de datos de entrenamiento, se realiza la propagación hacia adelante para calcular las predicciones del modelo y el valor de pérdida. Para un solo paso de tiempo $t$, se utilizan las ecuaciones matemáticas que describen la LSTM para calcular el estado oculto $H_t$, la unidad de memoria $C_t$, y las salidas de las puertas de olvido, entrada y salida. Esto se realiza iterativamente a lo largo de la secuencia de entrada.

3. **Cálculo de la Pérdida (Loss)**: La pérdida se calcula en función de las predicciones del modelo y los valores reales en el conjunto de datos de entrenamiento. Por ejemplo, para el Error Cuadrático Medio (MSE), la pérdida en el tiempo $t$ puede expresarse como:

   $$
   \mathcal{L}_t = \frac{1}{2}(Y_t - \hat{Y}_t)^2
   $$

   Donde $\mathcal{L}_t$ es la pérdida en el tiempo $t$, $Y_t$ es la salida real en el tiempo $t$, y $\hat{Y}_t$ es la salida predicha por la LSTM en el tiempo $t$.

4. **Retropropagación del Error (Backpropagation)**: Una vez que se ha calculado la pérdida, se utiliza el algoritmo de retropropagación del error (backpropagation) para calcular las derivadas parciales de la pérdida con respecto a todos los parámetros de la red, incluyendo los pesos y los sesgos. Las derivadas parciales se calculan usando la regla de la cadena.

   - **Ejemplo de Derivadas Parciales**:
   
     Para el cálculo de la derivada parcial de la pérdida $\mathcal{L}_t$ con respecto a un peso $W$, se utiliza la cadena de regla:

     $$
     \frac{\partial \mathcal{L}_t}{\partial W} = \frac{\partial \mathcal{L}_t}{\partial \hat{Y}_t} \cdot \frac{\partial \hat{Y}_t}{\partial H_t} \cdot \frac{\partial H_t}{\partial C_t} \cdot \frac{\partial C_t}{\partial W}
     $$

   Esto se repite para todos los pesos y sesgos de la LSTM.

5. **Actualización de Pesos (Optimización)**: Una vez que se han calculado las derivadas parciales, se utilizan algoritmos de optimización como el Descenso de Gradiente Estocástico (SGD), Adam, RMSProp, entre otros, para actualizar los pesos y los sesgos de la red en la dirección que minimiza la función de pérdida. La actualización se realiza utilizando la tasa de aprendizaje ($\alpha$) que controla la magnitud del ajuste.

   - **Ejemplo de Actualización de Pesos con SGD**:
   
     $$
     W \leftarrow W - \alpha \frac{\partial \mathcal{L}_t}{\partial W}
     $$

6. **Repetición del Proceso**: Los pasos 2 al 5 se repiten durante múltiples épocas, lo que significa iterar a través del conjunto de datos de entrenamiento varias veces. Esto permite que la LSTM ajuste sus parámetros gradualmente para aprender patrones en los datos.

7. **Validación y Evaluación**: Después del entrenamiento, se evalúa el rendimiento del modelo en un conjunto de datos de validación o prueba independiente para asegurarse de que pueda generalizar bien a datos no vistos. Se utilizan métricas de evaluación como el MSE, el Error Absoluto Medio (MAE) o coeficiente de correlación para medir la calidad de las predicciones.

8. **Ajuste de Hiperparámetros**: Durante el proceso de entrenamiento, es posible que sea necesario ajustar los hiperparámetros de la LSTM, como la tasa de aprendizaje, el tamaño de la capa oculta, la función de activación, etc., para optimizar el rendimiento del modelo.

Este proceso de entrenamiento iterativo permite que una LSTM aprenda patrones en los datos de series temporales y mejore su capacidad para hacer predicciones precisas. La elección adecuada de hiperparámetros, la función de pérdida y la arquitectura de la red son factores críticos para lograr un buen rendimiento en aplicaciones de series temporales.
